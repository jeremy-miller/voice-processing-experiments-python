# Librosa

This notebook contains code for the [librosa](https://librosa.org/doc/latest/index.html) library, including tutorials and additional experiments.

## Imports

```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
```

## Variables and Metadata

```python
target_sampling_rate = 16_000
mono = True

filename = "libri1"
filepath = librosa.example(filename)
audio_waveform, sampling_rate = librosa.load(filepath, sr=target_sampling_rate, mono=mono)

print(f"Filename: {filename}")
print(f"Filepath: {filepath}")
print(f"Sampling Rate: {sampling_rate}")
print(f"Mono: {mono}")
```

## Audio Playback

```python
Audio(data=audio_waveform, rate=sampling_rate)
```

## Waveform

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(audio_waveform, sr=sampling_rate)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

## Spectogram

```python
D = librosa.stft(audio_waveform)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
plt.figure()
librosa.display.specshow(S_db)
plt.colorbar()

fig, ax = plt.subplots()
img = librosa.display.specshow(S_db, ax=ax)
fig.colorbar(img, ax=ax)
```
